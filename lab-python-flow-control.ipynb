```python
products = ["t-shirt", "mug", "hat", "book", "keychain"]
inventory = {}

for product in products:
    quantity = input(f"Enter the quantity of '{product}' in stock:")
    inventory[product] = int(quantity)
    
customer_order = set()
```

```python
order = input("Do you want to order a product? (Yes/No)").lower()
```

```python
while order == "yes":
    item = input(f"What do you want to order, choose from {products}: ").lower()
    if item in products:
        customer_order.add(item)
        order = input("Do you want to add another product to your order? (Yes/No)").lower()
    else:
        print("Invalid product")
        item = input(f"What do you want to order, choose from {products}: ").lower()
    
```

```python
total_products_ordered = len(customer_order)
total_inventory = len(inventory)


print(f"Your order contains {total_products_ordered} items: {customer_order}")
```

## 3.

```python
for product in customer_order:
    if inventory[product] > 0:
        inventory[product] -= 1
        print(f"'{product}' has been shipped. Remaining stock: {inventory[product]}")
    else:
        print(f"Sorry, '{product}' is out of stock and cannot be shipped.")
    
```
